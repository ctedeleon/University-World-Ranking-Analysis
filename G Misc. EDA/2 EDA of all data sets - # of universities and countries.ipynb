### Create a new "region" column for the continents - to further categorize data

```python
#Change continent/region for new zealand and australia to oceania
#Create a list for each region (code used from Instacart task 4.10 jupyter notebook)
Asia_list = ['Pakistan','Qatar','Bangladesh','Indonesia','Jordan','Oman','China','Japan','South Korea','Taiwan','India','Israel','Turkey','Iran','Hong Kong','Saudi Arabia','Thailand','Malaysia','Singapore','United Arab Emirates','Lebanon']
```

```python
Africa_list = ['South Africa','Egypt','Uganda', 'Morocco', 'Nigeria', 'Ghana', 'Kenya']
NA_list = ['United States of America','Canada','Mexico','Puerto Rico']
SA_list = ['Brazil','Chile','Argentina','Colombia','Uruguay']
Europe_list = ['Ukraine', 'Latvia', 'Slovakia', 'Belarus','Republic of Ireland','United Kingdom','Germany','France','Italy','Spain','Netherlands','Switzerland','Austria','Sweden','Belgium','Finland','Greece','Poland','Ireland','Portugal','Denmark','Hungary','Norway','Czech Republic','Slovenia','Romania','Serbia','Cyprus','Slovak Republic','Iceland','Croatia','Bulgaria','Estonia','Lithuania','Russia','Russian Federation']
oceania_list = ['Australia', 'New Zealand']
```

```python
#Create a for-loop to search the "country" column
result = []
for value in Times["country"]:
    if value in Asia_list:
        result.append("Asia")
    elif value in Africa_list:
        result.append("Africa")
    elif value in NA_list: 
        result.append("North America")
    elif value in SA_list: 
        result.append("South America")
    elif value in Europe_list: 
        result.append("Europe")
    else: 
        result.append("Oceania")
```

### Apply the result to the times data frame

```python
#Create a new "region" column and set equal to the "result" from the for-loop 
Times['region_2'] = result
```
